```python
df_1000=pd.read_csv('../../../muse/data/mas/flights_1000.csv')
df_1000['_tid_'] = range(len(df_1000))
df_1000['is_dirty'] = 'False'
```

```python
unique_cities = list(df_1000['ORIGIN_CITY_NAME'].unique())
```

```python
i=1
row_cnt=len(df_1000)

while(i<=20):
    row_number = random.randint(0, row_cnt-1)
    gt_city= df_1000.at[row_number, 'ORIGIN_CITY_NAME']
    error_city = unique_cities[random.randint(0,len(unique_cities))-1]
    print(f"gt_city: {gt_city}")
    print(f"error_city: {error_city}")
    while(gt_city==error_city):
        gt_city= df_1000.at[row_number, 'ORIGIN_CITY_NAME']
        error_city = unique_cities[random.randint(0,len(unique_cities))-1]
        
    df_1000.at[row_number, 'ORIGIN_CITY_NAME'] = error_city
    df_1000.at[row_number,'is_dirty']='True'
    print(df_1000.at[row_number,'_tid_'])
    print('\n')
    i+=1
```
