```python
import pandas as pd
import json


data = []

with open('../../witan/data/plots.txt', 'r') as file:
    for line in file:
        data.append(json.loads(line.strip()))

df = pd.DataFrame(data)

pd.set_option('display.max_colwidth', None)

filtered_df = df[df['Genre'].str.contains('Action|Romance', case=False, na=False)]

# Display the filtered DataFrame
print(filtered_df)
```

```python
def get_class(x):
    if 'Action' in x:
        return 'action'
    elif 'Romance' in x:
        return 'romance'
    else:
        print('error')
```

```python
filtered_df.head()
```

```python
filtered_df['Genre_2class'] = filtered_df['Genre'].apply(get_class)
```

```python
filtered_df = filtered_df[['Plot', 'Genre_2class']]
```
