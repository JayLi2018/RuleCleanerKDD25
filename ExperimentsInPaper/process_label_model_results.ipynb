```python
import pickle
import pandas as pd
import glob
import sys
sys.path.append("/Users/chenjieli/Desktop/RuleCleaner/") 
```

```python
def process_results(directory):
    dicts = []
    for fname in glob.glob(directory+'*.pkl'):
        with open(fname,'rb') as f:
            d = pickle.load(f)['summary']
            dicts.append(d)
    return dicts
```

```python
if __name__ == '__main__':
    # directory = '../rulecleaner_src/test_other_label_model/'
    directory = '../rulecleaner_src/additional_other_label_models_majorities/'

    dicts = process_results(directory)
```

```python
df = pd.DataFrame(dicts)
```

```python
df['runtimes'].to_list()
```

```python
# Expand the dictionary column into separate columns
df_expanded = df["runtimes"].apply(pd.Series)

# If you want to merge it back with the original DataFrame (optional)
df_final = pd.concat([df, df_expanded], axis=1)
```
