```python
import pickle
import glob
import pandas as pd
import re
```

```python
res_gpt_as_labeler = {}
```

```python
for fname in glob.glob('*_results_from_gpt.pkl'):
    print(fname)
    try:
        with open(fname, 'rb') as f:
            df = pickle.load(f)
            print(list(df))
            accuracy = (df['preds_by_gpt'] == df['class']).mean()
            print(f"{fname} has accuracy {accuracy} by gpt")
            match = re.search(r"(?:old_)?([^_]+(?:_[^_]+)?)(?=_results)", fname)
            if match:
                extracted = match.group(1)
                print("extracted name: "+ extracted)  # Output: professor_teacher
                if('runtime' in df):
                    runtime = df['runtime'].unique()[0]
                else:
                    runtime = 'nan'
                res_gpt_as_labeler[extracted] = {'accuracy':accuracy, 'runtime':runtime, 'dataset_siz':len(df)}
    except:
        continue
```
